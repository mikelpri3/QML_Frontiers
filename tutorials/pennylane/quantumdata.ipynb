# Quantum Data

Pennylane have some pre-calculated data, for example the Hamiltonian of an Hydrogen molecule in a second

Good for simulate quantum sistems and discover new materials / medicines.

```python
import pennylane as qml
from pennylane import numpy as np


print("--- Available Dataset Categories ---")
# qchem, qspin, ...
available_data = qml.data.list_datasets()
print(list(available_data.keys()))


# Load Dataset(Quantum Chemistry)
print("\nLoading Hydrogen molecule (H2) dataset...")

# It downloads the data the first time, then caches it locally.
dataset = qml.data.load("qchem", molname="H2", basis="STO-3G", bondlength=0.742)[0] #specific basis set and bond length


# Inspect Quantum Data
print("\n--- Molecule Information ---")
print(f"Molecule name: {dataset.molname}")
print(f"Number of qubits required: {dataset.molecule.n_orbitals * 2}") # 2 qubits per orbital

print("\n--- Pre-calculated Results ---")
# Ground state energy (the absolute minimum energy we want our ML model to find)
print(f"True Ground State Energy: {dataset.vqe_energy:.6f} Hartrees")

print("\n--- The Hamiltonian (Objective Function) ---")
# This is the observable we would put inside qml.expval() to train our model
hamiltonian = dataset.hamiltonian
print(f"Hamiltonian terms: {len(hamiltonian.terms()[0])}")
# Print first 3 terms 
print(hamiltonian.terms()[1][:3])
```
